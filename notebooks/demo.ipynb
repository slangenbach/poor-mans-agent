# Demo

Demonstrating how to use the Poor man's agent


## Setup

```python
from poor_mans_agent import Agent
from poor_mans_agent.config import get_config
from poor_mans_agent.tools import read_url, search
```

```python
config = get_config()
```

## Usage

### Without tools

```python
agent = Agent(model=config.ai_model_name)
```

```python
response = agent.run("What are the basic building blocks to create an agent in Python?")
print(response)
```

### With tools

```python
agent = Agent(tools=[search, read_url])
```

```python
response = agent.run("What are the top 3 news in Germany today?")
print(response)
```

```python
response = agent.run(
    "Summarize the main idea of Build to Last blog at: https://www.fast.ai/posts/2025-10-30-build-to-last.html"
)
print(response)
```
